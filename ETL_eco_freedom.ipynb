# ECONOMIC FREEDOM DATA 2019

```python
# DEPENDENCIES AND SETUP

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

####  EXTRACTING DATA FROM KAGGLE

   ######  Data taken from url  https://www.kaggle.com/lewisduncan93/the-economic-freedom-index

```python
# Specyfying CSV path
eco_index = "./mini_project_economic_freedom.csv"
```

```python
# Reading csv file and creating dataframe

eco_df = pd.read_csv(eco_index)
eco_df.head()
```

#####  TRANSFORM DATA

```python
# Considering the columns needed

eco_index_new_df = eco_df[['CountryID', 'Country Name', 'Region', 'World Rank', 'Region Rank', '2019 Score', 'Labor Freedom', 'Population (Millions)', 'Unemployment (%)', 'Inflation (%)']].copy()
eco_index_new_df.head()
#eco_index_new_df
```

```python
# Deleteing unwanted rows by using dropna()

eco_df_clean = eco_index_new_df.dropna()
eco_df_clean.head()
#eco_df_clean
```

```python
# sort the dataframe by column name "World Bank"

eco_df_clean_sorted = eco_df_clean.sort_values('World Rank')
eco_df_clean_sorted.head()
```

```python
# Reseting the index 

eco_df_clean_reindex = eco_df_clean_sorted.reset_index(drop=True)
eco_df_clean_reindex.head()
#eco_df_clean_reindex
```

```python
# Consider top 5 countries based on "World Rank"

Top_Five_Condition = eco_df_clean_reindex['World Rank'] < 6
Top_Five = eco_df_clean_reindex[Top_Five_Condition]
Top_Five
```

```python
# Plotting the top Five countries economic freedom score of 2019

Top_Five_Eco = pd.pivot_table(Top_Five, index = 'Country Name', values = '2019 Score')
Top_Five_Eco.plot(kind = 'bar', color = "green", figsize=(10,6), fontsize = 15, rot=90)

#set title
plt.title('Top Five Countries - Economic Freedom Score 2019', fontsize = 15)
#set legend
plt.legend(bbox_to_anchor=(0, 1), loc='right', ncol=1, fontsize = 12)
#set yaxis label
plt.ylabel('Economic Freedom Score', fontsize = 15)
# set xaxis label
plt.xlabel('Countries', fontsize = 15)
# To save figure      
plt.savefig("Images/eco_top_five.png", bbox_inches="tight") 
# show figure
plt.show()
```

```python
# Consider bottom 5 countries based on "World Rank"

Bottom_Five_Condition = eco_df_clean_reindex['World Rank'] > 174
Bottom_Five = eco_df_clean_reindex[Bottom_Five_Condition]
Bottom_Five
```

```python
# Plotting the Bottom Five countries economic freedom score of 2019


Bottom_Five_Eco = pd.pivot_table(Bottom_Five, index = 'Country Name', values = '2019 Score')
Bottom_Five_Eco.plot(kind = 'bar', color = "coral", figsize=(10,6), fontsize = 15, rot=90)

#set title
plt.title('Bottom Five Countries - Economic Freedom Score 2019', fontsize = 15)
#set legend
plt.legend(bbox_to_anchor=(0, 1), loc='right', ncol=1, fontsize = 12)
#set yaxis label
plt.ylabel('Economic Freedom Score', fontsize = 15)
# set xaxis label
plt.xlabel('Countries', fontsize = 15)
# To save figure      
plt.savefig("Images/eco_bottom_five.png", bbox_inches="tight") 
# show figure
plt.show()
```

```python
# sort by column "Region" and taking columns "Country Name","Region","2019 Score","Labor Freedom"

eco_df_labor_freedom = eco_df_clean_reindex.sort_values("Region")[["Country Name","Region","2019 Score","Labor Freedom"]]
eco_df_labor_freedom.head()
```

```python
# Creating scatter plot based on 'Economic Freedom Score-2019','Labor Freedom' and "Region"

fig, ax = plt.subplots(figsize=(9,6))
#set margins
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
# set xaxis label
ax.set_xlabel('Labor Freedom', fontsize = 15)
# set yaxis label
ax.set_ylabel('Economic Freedom Score-2019', fontsize = 15)
# set title
ax.set_title('Economic Freedom Score-2019 Vs Labor Freedom', fontsize = 15)

# For loop is used  for ploting
# Do group by on dataframe "eco_df_labor_freedom" based on column "Region"
# loop through all groups one by one
for name, group in eco_df_labor_freedom.groupby('Region'):
    group.plot(x='Labor Freedom', y='2019 Score', ax=ax, marker='o',
               linestyle='', ms=10, label=name, fontsize = 15)

ax.legend(fontsize = 10)
# save figure
plt.savefig("Images/scatter_plot.png", bbox_inches="tight")
# show figure
plt.show()
```
